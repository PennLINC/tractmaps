## Load data

```python
# Directory containing all the raw folders
root_dir = f'/Users/joelleba/Data/tractmaps/neurosynth/raw/neurosynth'

# Directory containing glasser parcellated neurosynth terms generated with parcellate_neurosynth.ipynb
out_dir = f'/Users/joelleba/PennLINC/tractmaps/data/neurosynth_annotations/glasser/'
```

```python
# Create the cogatl_maps dictionary using dictionary comprehension
print('Loading association maps...') # these were already saved to a pickle for analyses
glasser_cogatl_maps = tm_utils.load_data(f'{out_dir}/glasser_neurosynth_123terms_dict.pickle')

# Create the cogatl_maps dictionary using dictionary comprehension
print('Loading FDR corrected association maps...') # these aren't used for analyses so I'm loading them here separately
cogatl_maps_fdr = {
    term_name: {
        'annotation': ('cognitive atlas', term_name, 'MNI152'), # space (last term) will be used in glasserize() to parcellate data
        'map': os.path.join(root_dir, term_name, f'z_desc-association_level-voxel_corr-FDR_method-indep.nii.gz')
    }
    for term_name in os.listdir(root_dir)
    if os.path.isdir(os.path.join(root_dir, term_name))
}

print(f'Loaded {len(cogatl_maps_fdr)} cognitive terms into the dictionary.')

# generate glasser parcellated neurosynth maps 
glasser_cogatl_maps_fdr = tm_parcellate.glasserize(cortical_maps = cogatl_maps_fdr, zscore = False) # z-scoring not applied as NS maps are already z-scored
print(f'Applied Glasser parcellation!')
```

## Compare current data (z-desc association and FDR corrected) to Audrey's data 

```python
### df 1: current ns data ###

# create the df
df_glasser_cogatl_maps = pd.DataFrame(glasser_cogatl_maps) # non FDR corrected
# df_glasser_cogatl_maps = pd.DataFrame(glasser_cogatl_maps_fdr) # FDR corrected
df_glasser_cogatl_maps['regionID'] = range(1, len(df_glasser_cogatl_maps) + 1)
sorted_columns = sorted(df_glasser_cogatl_maps.columns.difference(['regionID'])) # Sort the cognitive terms columns alphabetically
df_glasser_cogatl_maps = pd.concat([df_glasser_cogatl_maps['regionID'], df_glasser_cogatl_maps[sorted_columns]], axis = 1) # reorder columns

# remove glasser suffix
suffix_to_remove = '_glasser'
for col_name in df_glasser_cogatl_maps.columns:
    if col_name.endswith(suffix_to_remove):
        # Remove the suffix and update the column name
        new_col_name = col_name[:-len(suffix_to_remove)]
        df_glasser_cogatl_maps.rename(columns = {col_name: new_col_name}, inplace = True)
        
# load df with tracts probs and glasser labels 
tracts_regs_ids = pd.read_csv(f'/Users/joelleba/PennLINC/tractmaps/data/tracts/glasser/glasser_tracts.csv')
tracts_regs_ids.rename(columns={'parcel_name': 'regionName'}, inplace = True)
regs_ids = tracts_regs_ids[['regionID', 'regionName']]

# add glasser labels to ns data
current_ns_df = pd.merge(regs_ids, df_glasser_cogatl_maps)

# remove region ID
current_ns_df.pop('regionID')

# sort regions
current_ns_df = current_ns_df.sort_values(by = 'regionName')
current_ns_df.pop('regionName')

### df 2: old ns data ####

# load old ns data  
old_ns_df = pd.read_csv('/Users/joelleba/Downloads/atl-glasser360_neurosynth.csv')

# place the hemisphere (L_ or R_) as a prefix, not a suffix in region labels
old_ns_df['regionName'] = old_ns_df['regionName'].apply(lambda x: f"{x[-1]}_{x[:-2]}")

# Sort the rows based on the 'regionName' column
old_ns_df = old_ns_df.sort_values(by = 'regionName')
old_ns_df.pop('regionName')
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Initialize an empty DataFrame for the correlation matrix
correlation_matrix = pd.DataFrame(index = current_ns_df.columns, columns = old_ns_df.columns)

# Compute the correlations
for col1 in current_ns_df.columns:
    for col2 in old_ns_df.columns:
        correlation_matrix.loc[col1, col2] = pearsonr(current_ns_df[col1], old_ns_df[col2])[0]
        
```

```python
# Set up the matplotlib figure
plt.figure(figsize=(12, 8))

# Generate a heatmap with the correlation matrix
sns.heatmap(correlation_matrix.astype(float), cmap='coolwarm', linewidths=0.5,
           center = 0, vmax = 1, vmin = -1)

# Show the plot
plt.title('Correlation Heatmap between current nsdata and old nsdata')
plt.show()
```

```python
# get correlations of the diagonal (i.e. corresponding terms in current vs old data)
corr_min = np.min(np.diag(correlation_matrix))
corr_mean = np.mean(np.diag(correlation_matrix))
corr_max = np.max(np.diag(correlation_matrix))
print(f'Correlation \nmin: {round(corr_min, 2)} \nmean: {round(corr_mean, 2)} \nmax: {round(corr_max, 2)}')
```
