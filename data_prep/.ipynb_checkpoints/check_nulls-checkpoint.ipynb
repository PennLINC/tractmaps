# Check spun data

This notebook performs a couple of sanity checks to ensure that the generated null data is as expected. 

Here are the things we check for: 

* neurosynth data: spun data are different from original, but with a relatively similar distribution
* tract data: 
    + Spun data are different from original. Connection probabilities are shuffled, but remain on the same hemisphere as the tract. In other words, no NaNs from the other hemisphere ever end up in the tract's hemisphere. And vice versa, proportion data never end up on the opposite hemisphere. 
    + Spun data have a relatively similar distribution compared to the empirical tract connection data.

```python
import os
import pandas as pd
import numpy as np
import sys
sys.path.append('..') # Move one directory up to access 'code' directory
from utils import tm_utils
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
### NEUROSYNTH 124 COGNITIVE TERMS ###
analysis = 'check_nulls'
data_root = '/Users/joelleba/PennLINC/tractmaps/data/neurosynth_annotations/'
data = pd.read_csv(f'{data_root}/glasser/glasser_tracts_neurosynth_123terms.csv')
nsdata = data.iloc[:, 59:]
tractdata = data.filter(regex='left|right')
```

```python
# set up inputs
nspins = 10000
root = '/Users/joelleba/PennLINC/tractmaps'
```

```python
# create results directory if it doesn't yet exist
results_dir = f'{root}/results/{analysis}/'
if not os.path.exists(results_dir):
    os.makedirs(results_dir)
    print(f"Folder '{results_dir}' created.")
else:
    print(f"Folder '{results_dir}' already exists.")
```

## Plot spun data

```python
tract_spins = tm_utils.load_data(f'{root}/data/nulls/tracts/tracts_{nspins}spins.pickle')

for hemisphere in ['left', 'right']:
    
    print(f'Running for {hemisphere} hemisphere...')
    hem_tractdata = tractdata.filter(regex = hemisphere)
    
    if hemisphere == 'left':
        hem_tractdata, hem_nsdata = hem_tractdata.iloc[:180, ], nsdata.iloc[:180, ]
        tract_names, term_names = hem_tractdata.columns, hem_nsdata.columns
        hem_spins = {tract: spin_vals[:180, :100] for tract, spin_vals in tract_spins.items() if tract in tract_names}

    elif hemisphere == 'right':
        hem_tractdata, hem_nsdata = hem_tractdata.iloc[180:, ], nsdata.iloc[180:, ]
        tract_names, term_names = hem_tractdata.columns, hem_nsdata.columns
        hem_spins = {tract: spin_vals[180:, :1000] for tract, spin_vals in tract_spins.items() if tract in tract_names}
        
    # Set up the grid
    num_tracts = len(hem_spins)
    cols = 5
    rows = (num_tracts + cols - 1) // cols  # Calculate rows needed
    fig, axes = plt.subplots(rows, cols, figsize=(15, rows * 3))

    # Flatten the axes array for easy iteration
    axes = axes.flatten()

   # Plot each tract's distributions
    for idx, (tract_name, null_data) in enumerate(hem_spins.items()):
        empirical_data = hem_tractdata[tract_name]

        # Convert to NumPy array
        null_data = np.array(null_data)
        empirical_data = np.array(empirical_data)

        # Plot each column (spin) of null data as a separate line
        for spin in range(null_data.shape[1]):
            sns.kdeplot(null_data[:, spin], color='royalblue', alpha=0.2, ax=axes[idx])

        # Plot empirical distribution
        sns.kdeplot(empirical_data, color='red', linewidth=2, label='Empirical', ax=axes[idx])
        
        # Add a custom legend entry for the null distributions
        axes[idx].plot([], [], color='royalblue', alpha = 0.5, label='Null Distributions')

        # Titles and labels
        axes[idx].set_title(tract_name)
        axes[idx].set_xlabel('Values')
        axes[idx].set_ylabel('Density')

        # Add legend only for the first subplot
        if idx == 0:
            axes[idx].legend()
            
    # Remove empty subplots if any
    for i in range(len(hem_spins), len(axes)):
        fig.delaxes(axes[i])
    fig.suptitle('Comparison of empirical vs null distributions of tracts (0 to 1 proportions) \n\n')

    # Adjust layout for better spacing
    plt.tight_layout()
    plt.savefig(f'{results_dir}/{hemisphere}_tractnulls.png', bbox_inches='tight', dpi=300, transparent=False)
    plt.show()
```

```python
term_spins = tm_utils.load_data(f'{root}/data/nulls/neurosynth/neurosynth_{nspins}spins.pickle')

for hemisphere in ['left', 'right']:
    
    print(f'Running for {hemisphere} hemisphere...')
    hem_tractdata = tractdata.filter(regex = hemisphere)
    
    if hemisphere == 'left':
        hem_tractdata, hem_nsdata = hem_tractdata.iloc[:180, :20], nsdata.iloc[:180, :20] # plot only first 20 terms
        tract_names, term_names = hem_tractdata.columns, hem_nsdata.columns
        hem_spins = {term: spin_vals[:180, :100] for term, spin_vals in term_spins.items() if term in term_names}

    elif hemisphere == 'right':
        hem_tractdata, hem_nsdata = hem_tractdata.iloc[180:, :20], nsdata.iloc[180:, :20] # plot only first 20 terms
        tract_names, term_names = hem_tractdata.columns, hem_nsdata.columns
        hem_spins = {term: spin_vals[180:, :100] for term, spin_vals in term_spins.items() if term in term_names}
        
        
    # Set up the grid
    num_terms = len(hem_spins)
    cols = 5
    rows = (num_terms + cols - 1) // cols  # Calculate rows needed
    fig, axes = plt.subplots(rows, cols, figsize=(15, rows * 3))

    # Flatten the axes array for easy iteration
    axes = axes.flatten()

   # Plot each tract's distributions
    for idx, (term_name, null_data) in enumerate(hem_spins.items()):
        empirical_data = hem_nsdata[term_name]

        # Convert to NumPy array
        null_data = np.array(null_data)
        empirical_data = np.array(empirical_data)

        # Plot each column (spin) of null data as a separate line
        for spin in range(null_data.shape[1]):
            sns.kdeplot(null_data[:, spin], color='royalblue', alpha=0.2, ax=axes[idx])

        # Plot empirical distribution
        sns.kdeplot(empirical_data, color='red', linewidth=2, label='Empirical', ax=axes[idx])
        
        # Add a custom legend entry for the null distributions
        axes[idx].plot([], [], color='royalblue', alpha = 0.5, label='Null Distributions')

        # Titles and labels
        axes[idx].set_title(term_name)
        axes[idx].set_xlabel('Values')
        axes[idx].set_ylabel('Density')

        # Add legend only for the first subplot
        if idx == 0:
            axes[idx].legend()
            
    # Remove empty subplots if any
    for i in range(len(hem_spins), len(axes)):
        fig.delaxes(axes[i])

    fig.suptitle('Comparison of empirical vs null distributions of cognitive terms (z-scored) \n\n')
    
    # Adjust layout for better spacing
    plt.tight_layout()
    plt.savefig(f'{results_dir}/{hemisphere}_termnulls.png', bbox_inches='tight', dpi=300, transparent=False)
    plt.show()
```

```python
tract_spins = tm_utils.load_data(f'{root}/data/nulls/tracts/tracts_{nspins}spins.pickle')

for hemisphere in ['left', 'right']:
    
    print(f'Running for {hemisphere} hemisphere...')
    hem_tractdata = tractdata.filter(regex = hemisphere)
    
    if hemisphere == 'left':
        hem_tractdata, hem_nsdata = hem_tractdata.iloc[:180, ], nsdata.iloc[:180, ]
        tract_names, term_names = hem_tractdata.columns, hem_nsdata.columns
        hem_spins = {tract: spin_vals[:180, :] for tract, spin_vals in tract_spins.items() if tract in tract_names}

    elif hemisphere == 'right':
        hem_tractdata, hem_nsdata = hem_tractdata.iloc[180:, ], nsdata.iloc[180:, ]
        tract_names, term_names = hem_tractdata.columns, hem_nsdata.columns
        hem_spins = {tract: spin_vals[180:, :] for tract, spin_vals in tract_spins.items() if tract in tract_names}
        
    # Initialize lists to store proportions
    tract_names = []
    empirical_proportions = []
    null_proportions_list = []

    # Calculate the proportion of non-zero values
    for tract_name, null_data in hem_spins.items():
        empirical_data = hem_tractdata[tract_name]

        # Calculate proportion of non-zero values for empirical data
        empirical_non_zero = np.count_nonzero(empirical_data) / empirical_data.size

        # Calculate proportion of non-zero values for each column in null data
        null_non_zero = np.count_nonzero(null_data, axis=0) / null_data.shape[0]

        # Store the results
        tract_names.append(tract_name)
        empirical_proportions.append(empirical_non_zero)
        null_proportions_list.append(null_non_zero)

    # Sort by empirical proportions
    sorted_indices = np.argsort(empirical_proportions)
    tract_names = [tract_names[i] for i in sorted_indices]
    empirical_proportions = [empirical_proportions[i] for i in sorted_indices]
    null_proportions_list = [null_proportions_list[i] for i in sorted_indices]

    # Prepare the data for Seaborn boxplot
    null_df = pd.DataFrame(null_proportions_list).T.melt(var_name='Tract', value_name='Proportion')
    null_df['Tract'] = null_df['Tract'].map(dict(enumerate(tract_names)))

    # Plotting
    fig, ax = plt.subplots(figsize=(8, 5))

    # Seaborn boxplot for null proportions
    sns.boxplot(x='Tract', y='Proportion', data=null_df, ax=ax, color='white', width = 0.4,
                boxprops=dict(alpha=0.5), whiskerprops=dict(color='grey'), showcaps=False, showfliers=False)

    # Scatter plot for empirical proportions
    ax.scatter(np.arange(len(tract_names)), empirical_proportions, color='lightcoral', label='Empirical', zorder=3)

    # Customize plot
    ax.set_xlabel('Tracts')
    ax.set_ylabel('Proportion of Non-Zero Values')
    ax.set_title('Comparison of Non-Zero Value Proportions in Empirical vs Null Data')
    ax.set_xticks(np.arange(len(tract_names)))
    ax.set_xticklabels(tract_names, rotation=45, ha='right')
    ax.legend()

    # Display plot
    plt.tight_layout()
    plt.savefig(f'{results_dir}/{hemisphere}_tracts_nonzeroproportions.png', bbox_inches='tight', dpi=300, transparent=False)
    plt.show()
```
